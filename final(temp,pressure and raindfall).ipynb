```python
#Importing Libraries
#For rainfall
import pandas as pd
```

```python
#Loading Data from CSV file
dataset = pd.read_csv('Weather.csv')
```

```python
dataset.head()
```

```python
#Check if table has missing values
pd.isnull(dataset).any(1).nonzero()[0]
```

```python
dataset.head()
```

```python
#Drop rows that have missing values
dataset.drop(pd.isnull(dataset).any(1).nonzero()[0], inplace = True)
```

```python
#Breaking down Independent and Dependent variables
R = dataset.iloc[: , 1:4].values #upperbound is omitted
a = dataset.iloc[:, 4].values.astype(int)
y_temp = dataset.iloc[:, 4].values
#R=dataset.drop(['RAIN'],axis=1)
#a=dataset.drop(['PRCP'],axis=1)
```

```python
type(R)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
R_train, R_test, a_train, a_test = train_test_split(R, a, test_size = 0.20, random_state = 0)
```

```python
R_train
```

```python
type(R_train)
```

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler(feature_range=(0,1))
R_train = ms.fit_transform(R_train)
R_test = ms.transform(R_test)
```

```python
#Fitting the Classifier
from sklearn.svm import SVC
classifier = SVC(kernel='linear', random_state = 0)
classifier.fit(R_train, a_train)
```

```python
type(R_train)
```

```python
#Predicting rain/no
a_pred = classifier.predict(R_test)
```

```python
#Analysing our model
from sklearn.model_selection import cross_val_score
scores = cross_val_score(classifier, R_train, a_train, cv=5)
scores.mean() 
```

```python
#Analysing our model
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(a_test, a_pred)
```

```python
#Printing Results
print('Our model has correctly predicted that there will be no rain for', cm[0,0] , 'days.')
print('Our model has correctly predicted that there will be rain for', cm[1,1] , 'days.')

```
